```python
import pandas as pd
import numpy as np
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('data2021.csv', index_col=0)
data
```

```python
rating = pd.read_csv('rating2021.csv', header=1).iloc[1:-2, :]
rating
```

```python
rating_selected = rating.loc[:, ['Ф.И.О.', 'рейтинг']].rename(columns={'Ф.И.О.': 'ФИО'})
rating_selected
```

```python
rating_selected_dropna = rating_selected.dropna()
rating_selected_dropna
```

```python
data_selected = data.loc[:, ['ФИО', 'Блок 1 (7 семестр)', 'Блок 2 (7 семестр)', 'Блок 3 (8 семестр)']]
data_selected
```

```python
merge_df = data_selected.merge(rating_selected_dropna, on='ФИО').set_index('ФИО')
merge_df
```

```python
merge_df = merge_df.replace('Дизайн (Системы постпроцессинга )- Газизов Р.', 'Дизайн (Системы постпроцессинга)- КостюкД.')
```

```python
def get_cat(x):
    return float(x.replace(',', '.'))
```

```python
merge_df['рейтинг'] = merge_df['рейтинг'].apply(get_cat)
merge_df
```

```python
X = merge_df.drop(columns=['Блок 3 (8 семестр)'])
y = merge_df['Блок 3 (8 семестр)']
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=23)
```

```python
clf = CatBoostClassifier(
    iterations=100, 
    learning_rate=0.1, 
    max_depth=6,
    random_seed=23,
    eval_metric='Accuracy'
)


clf.fit(X_train, y_train, 
        cat_features=[0, 1], 
        eval_set=(X_val, y_val), 
        verbose=True)
```
